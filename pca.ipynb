```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv(r"C:\Users\Kuberan\Desktop\Fitness_data.csv")
features = data[['age', 'height_cm', 'weight_kg', 'daily_steps', 'avg_heart_rate']]
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_features)
```

```python
pc_df = pd.DataFrame(data=principal_components, columns=['Principal Component 1', 'Principal Component 2'])
plt.scatter(pc_df['Principal Component 1'], pc_df['Principal Component 2'])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Fitness Data')
plt.show()
```
